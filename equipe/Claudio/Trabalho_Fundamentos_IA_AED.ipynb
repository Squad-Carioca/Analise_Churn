```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
import os
```

```python
#fonte: https://www.kaggle.com/datasets/mnassrib/telecom-churn-datasets
DATA_PATH = os.path.join(".." ,"..","data", "churn-bigml-80.csv") 

df=pd.read_csv(DATA_PATH)
```

```python
df.shape
```

```python
rename_map = {
    'State': 'Estado',
    'Account length': 'TempoConta',
    'Area code': 'CodigoArea',
    'International plan': 'PlanoInternacional',
    'Voice mail plan': 'PlanoCorreioVoz',
    'Number vmail messages': 'NumerodeMensagensdeVoz',
    'Total day minutes': 'TotalMinutosDia',
    'Total day calls': 'TotalChamadasDia',
    'Total day charge': 'CustoTotalDia',
    'Total eve minutes': 'TotalMinutosTardeNoite',
    'Total eve calls': 'TotalChamadasTardeNoite',
    'Total eve charge': 'CustoTotalTardeNoite',
    'Total night minutes': 'TotalMinutosNoturno',
    'Total night calls': 'TotalChamadasNoturno',
    'Total night charge': 'CustoTotalNoturno',
    'Total intl minutes': 'TotalMinutosInternacionais',
    'Total intl calls': 'TotalChamadasInternacionais',
    'Total intl charge': 'CustoTotalInternacional',
    'Customer service calls': 'ChamadasSuporte',
    'Churn': 'Cancelamento (Churn)'
}

df.rename(columns=rename_map, inplace=True)

display(df.head().T)
```

##Dicionário de Dados

| Campo Original (Inglês) | Tradução (Português) | Tipo de Dado (Inferido) | Descrição e Exemplos |
| :--- | :--- | :--- | :--- |
| **State** | **Estado** | Categórico/String | O estado de residência do cliente (e.g., LA, IN, NY, AZ). |
| **Account length** | **TempoConta** | Numérico (Integer) | O número de dias que o cliente tem a conta (e.g., 117, 65, 161). |
| **Area code** | **CodigoArea** | Categórico/Numérico (Integer) | O código de área telefônico do cliente (e.g., 408, 415, 510). |
| **International plan** | **PlanoInternacional** | Categórico (Boolean) | Indica se o cliente possui um plano internacional ativo (Yes/No). |
| **Voice mail plan** | **PlanoCorreioVoz** | Categórico (Boolean) | Indica se o cliente possui um plano de correio de voz ativo (Yes/No). |
| **Number vmail messages** | **NumerodeMensagensdeVoz** | Numérico (Integer) | O número de mensagens de correio de voz (e.g., 0, 22, 34). |
| **Total day minutes** | **TotalMinutosDia** | Numérico (Float) | Total de minutos de chamadas realizadas durante o dia (e.g., 184.5, 129.1). |
| **Total day calls** | **TotalChamadasDia** | Numérico (Integer) | Número total de chamadas realizadas durante o dia (e.g., 97, 137). |
| **Total day charge** | **CustoTotalDia** | Numérico (Float) | Custo total das chamadas realizadas durante o dia (e.g., 31.37, 21.95). |
| **Total eve minutes** | **TotalMinutosTardeNoite** | Numérico (Float) | Total de minutos de chamadas realizadas durante a noite/tarde (evening) (e.g., 351.6, 228.5). |
| **Total eve calls** | **TotalChamadasTardeNoite** | Numérico (Integer) | Número total de chamadas realizadas durante a noite/tarde (e.g., 80, 83). |
| **Total eve charge** | **CustoTotalTardeNoite** | Numérico (Float) | Custo total das chamadas realizadas durante a noite/tarde (e.g., 29.89, 19.42). |
| **Total night minutes** | **TotalMinutosNoturno** | Numérico (Float) | Total de minutos de chamadas realizadas durante a madrugada (e.g., 215.8, 208.8). |
| **Total night calls** | **TotalChamadasNoturno** | Numérico (Integer) | Número total de chamadas realizadas durante a madrugada (e.g., 90, 111). |
| **Total night charge** | **CustoTotalNoturno** | Numérico (Float) | Custo total das chamadas realizadas durante a madrugada (e.g., 9.71, 9.4). |
| **Total intl minutes** | **TotalMinutosInternacionais** | Numérico (Float) | Total de minutos de chamadas internacionais (e.g., 8.7, 12.7). |
| **Total intl calls** | **TotalChamadasInternacionais** | Numérico (Integer) | Número total de chamadas internacionais (e.g., 4, 6). |
| **Total intl charge** | **CustoTotalInternacional** | Numérico (Float) | Custo total das chamadas internacionais (e.g., 2.35, 3.43). |
| **Customer service calls** | ** ChamadasSuporte ** | Numérico (Integer) | Número de chamadas para o serviço ao cliente (e.g., 1, 4). |
| **Churn** | **Cancelamento (Churn)** | Categórico (Boolean) | Variável alvo que indica se o cliente cancelou o serviço (True/False). |

```python
fig,ax = plt.subplots(nrows=4,ncols=2,figsize=(30,30))

df['Cancelamento (Churn)'].value_counts().plot(kind='pie',ax=ax[0][0],autopct='%1.1f%%',startangle=90, legend=True)
ax[0][0].set_ylabel('')  # Remove o rótulo do eixo Y
ax[0][0].set_title('Cancelamentos')

#sb.countplot(data=df,x='Cancelamento (Churn)',ax=ax[0][0])
# ax[0][0].set_ylabel('')  # Remove o rótulo do eixo Y
# ax[0][0].set_title('Cancelamentos')

#distribuição de cancelamentos por estado
sb.countplot(data=df, x='Estado',hue='Cancelamento (Churn)',ax=ax[0][1])
ax[0][1].set_title('Cancelamentos por Estado')

#distribuição de cancelamentos por código de área
sb.countplot(data=df, x='CodigoArea',hue='Cancelamento (Churn)',ax=ax[1][0])
ax[1][0].set_title('Cancelamentos por CodigoArea')

#distribuição de cancelamentos por Plano Internacional
sb.countplot(data=df, x='PlanoInternacional',hue='Cancelamento (Churn)',ax=ax[1][1])
ax[1][1].set_title('Cancelamentos por Plano Internacional')

#distribuição de cancelamentos por Correio de Voz
sb.countplot(data=df, x='PlanoCorreioVoz',hue='Cancelamento (Churn)',ax=ax[2][0])
ax[2][0].set_title('Cancelamentos por Plano Correio Voz')


#por faixa de uso
max_val = df['TotalMinutosDia'].max()
# Cria as faixas (0, 50, 100, ..., até o máximo arredondado para cima)
bins = list(range(0, int(max_val + 50), 50))
# Cria uma nova coluna com as faixas
df['Faixa_50'] = pd.cut(df['TotalMinutosDia'], bins=bins, include_lowest=True)

sb.countplot(data=df, x='Faixa_50',hue='Cancelamento (Churn)',ax=ax[2][1])
ax[2][1].set_title('Cancelamentos por Faixa_50')


# #por faixa de Chamadas ao Suporte
# max_val = df['ChamadasSuporte'].max()
# # Cria as faixas
# bins = list(range(0, int(max_val + 2), 2))
# # Cria uma nova coluna com as faixas
# df['Faixa_5'] = pd.cut(df['ChamadasSuporte'], bins=bins, include_lowest=True)

sb.countplot(data=df, x='ChamadasSuporte',hue='Cancelamento (Churn)',ax=ax[3][0])
ax[3][0].set_title('Cancelamentos por ChamadasSuporte')


#por faixa de Consumo Diário
max_val = df['CustoTotalDia'].max()
# Cria as faixas
bins = list(range(0, int(max_val + 10), 10))
# Cria uma nova coluna com as faixas
df['Faixa_custo_diario'] = pd.cut(df['CustoTotalDia'], bins=bins, include_lowest=True)

sb.countplot(data=df, x='Faixa_custo_diario',hue='Cancelamento (Churn)',ax=ax[3][1])
ax[3][1].set_title('Cancelamentos por CustoTotalDia')

plt.show()

#df.plot(kind='pie',df['Cancelamento (Churn)'])
```
