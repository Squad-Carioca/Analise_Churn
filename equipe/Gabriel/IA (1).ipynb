```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/churn-bigml.csv")
```

```python
print("O tamanho do arquivo é:", df.shape)

```

```python
display(df.head())
```

```python
rename_map = {
    'State': 'Estado',
    'Account length': 'TempoConta',
    'Area code': 'CodigoArea',
    'International plan': 'PlanoInternacional',
    'Voice mail plan': 'CaixaPostal',
    'Number vmail messages': 'NumerodeMensagensdeVoz',
    'Total day minutes': 'MinutosDiurnos',
    'Total day calls': 'ChamadasDiurnas',
    'Total day charge': 'CustoDiurno',
    'Total eve minutes': 'MinutosTarde',
    'Total eve calls': 'ChamadasTarde',
    'Total eve charge': 'CustoTarde',
    'Total night minutes': 'MinutosNoturnos',
    'Total night calls': 'ChamadasNoturnas',
    'Total night charge': 'CustoNoturno',
    'Total intl minutes': 'MinutosInternacionais',
    'Total intl calls': 'ChamadasInternacionais',
    'Total intl charge': 'CustoInternacional',
    'Customer service calls': 'ChamadasSuporte',
    'Churn': 'Cancelamento(churn)'
}

df.rename(columns=rename_map, inplace=True)

display(df.head())
```

```python
df['PlanoInternacional'] = df['PlanoInternacional'].replace({
    'Yes': 'Possui plano internacional',
    'No': 'Não possui plano internacional'
})

display(df.head())
```

```python
# Substituir valores booleanos
df['Cancelamento(churn)'] = df['Cancelamento(churn)'].replace({
    False: 'Não cancelou',
    True: 'Cancelou'
})
display(df.head())
# Conferir
df['Cancelamento(churn)'].value_counts()
```

Observações principais da análise exploratória

1 - quantos clientes canceleram m o plano?

2 - Quantidade de ligações para o suporte

3 - Possui plano internacional?

4 - Total de minutos em ligação

Passo a passo para criar o gráfico

1- plotar as dimensões do gráfico

2- Plotar o tipo de gráfico e colocar as cores do gráfico

3- Colocar o titulo do gráfico

4- Colocar o tituto do eixo X

5- Colocar o titulo do eixo Y

```python
#gráfico de colunas para mostrar qtde de cancelamentos
plt.figure(figsize=(6,4))
sns.countplot(x='Cancelamento(churn)', data=df, palette='coolwarm') #tipo de gráfico e coluna que usa como base
plt.title('Distribuição da Variável Alvo: Cancelamento') #tituos
plt.xlabel('Churn (Cancelou)') #Rótulo do eixo X
plt.ylabel('Quantidade de Clientes') #Rótulo do eixo Y
plt.show() #Plotar (mostrar) gráfico
contagem = df['Cancelamento(churn)'].value_counts() # Gera a contagem de cancelamentos (0 = não cancelou, 1 = cancelou)

# Cria o gráfico de pizza que mostra o percentual de cancelamentos
plt.figure(figsize=(6,6))
plt.pie(
    contagem,
    labels=["Não cancelou", "Cancelou"],
    autopct='%1.1f%%',
    shadow=True,
    startangle=90,
    colors=['#66b3ff', '#ff9999']
)
plt.title('Percentual de Cancelamento')
plt.axis('equal')  # deixa o gráfico redondo
plt.show()


#gráfico de colunas para mostrar o impacto do plano internacional no cancelamento
plt.figure(figsize=(6,4))
ax=sns.countplot(x="PlanoInternacional", hue="Cancelamento(churn)", data=df, palette='Set2')
plt.title("Cancelamento por plano internacional")
plt.xlabel("PlanoInternacional")
plt.ylabel("Cancelamento(churn)")
plt.show() #Plotar (mostrar) gráfico
contagem = df['Cancelamento(churn)'].value_counts()
plt.show()


# Total de ligações para o suporte
total_suporte = df['ChamadasSuporte'].sum()
# Criar gráfico de barra única
plt.figure(figsize=(5,3))
plt.bar(x=["Total"], height=[total_suporte], color="#66b3ff")
plt.title('Total de Ligações para o Suporte')
plt.ylabel('Quantidade de Ligações')
# Exibe o número acima da barra
plt.text(0, total_suporte + 10, f'{total_suporte:,}',
         ha='center', fontsize=12, fontweight='bold', color='black')
plt.show()


# Total de ligações para o suporte x cancelamentos
plt.figure(figsize=(6,4))
sns.boxplot(x="Cancelamento(churn)", y="ChamadasSuporte", data=df, palette="Pastel1")
plt.title("Cancelamentos x Ligações para suporte")
plt.xlabel("Cancelamento(churn)")
plt.ylabel("ChamadasSuporte")
plt.show()

#Total de minutos diurnos usados x cancelamento
plt.figure(figsize=(6,4))
sns.histplot(data=df, x="MinutosDiurnos", hue="Cancelamento(churn)", kde=True, palette="coolwarm", bins=30)
plt.title("Distribuição dos Minutos Diurnos por Status de Churn")
plt.xlabel("Minutos de Ligação Diurnos")
plt.ylabel("Contagem")
plt.show()
```
